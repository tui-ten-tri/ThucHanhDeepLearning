```python
#2001230980 - Bùi Quốc Trí - Lab1_DeepLearning
import numpy as np

# =========================================
# 1. TẠO LIST PYTHON
# =========================================

my_list = [[1., 2., 3.], [4., 5., 6.]]
my_list2 = [[1., 2., 3.]]
my_list3 = [[1., 2., 3.]]

print("Python List:")
print(my_list)

# =========================================
# 2. CHUYỂN SANG NUMPY ARRAY
# =========================================

np_arr = np.array(my_list)
np_arr2 = np.array(my_list2)
np_arr3 = np.array(my_list3)

print("\nNumPy Array:")
print(np_arr)

# =========================================
# 3. SHAPE
# =========================================

print("\nShape:")
print(np_arr.shape)

# =========================================
# 4. RESHAPE
# =========================================

print("\nReshape 2x3 -> 3x2:")
print(np_arr.reshape(3, 2))

# =========================================
# 5. DTYPE
# =========================================

print("\nDtype:")
print(np_arr.dtype)

# =========================================
# 6. ONES
# =========================================

print("\nnp.ones((2,2)):")
print(np.ones(shape=(2, 2)))

# =========================================
# 7. FLATTEN
# =========================================

print("\nFlatten:")
print(np_arr.flatten())

# =========================================
# 8. HSTACK
# =========================================

print("\nHorizontal Stack:")
print(np.hstack((np_arr3, np_arr2)))

# =========================================
# 9. VSTACK
# =========================================

print("\nVertical Stack:")
print(np.vstack((np_arr3, np_arr2)))

# =========================================
# 10. RANDOM RAND
# =========================================

print("\nRandom Rand (3,2,3):")
print(np.random.rand(3, 2, 3))

# =========================================
# 11. RANDOM NORMAL
# =========================================

print("\nRandom Normal:")
print(np.random.normal(5, 1, 10))

# =========================================
# 12. RANDOM CHOICE
# =========================================

print("\nRandom Choice:")
print(np.random.choice([1, 2, 3], size=(5, 5)))

# =========================================
# 13. MATRIX
# =========================================

A = np.matrix(np.ones((5, 5)))

print("\nMatrix A:")
print(A)

# Chuyển sang array để chỉnh sửa
B = np.array(A)

# Sửa hàng thứ 2
B[2] = 2

print("\nModified Array B:")
print(B)

# =========================================
# 14. ARANGE
# =========================================

print("\nArange:")
print(np.arange(1, 20, 3))

# =========================================
# 15. LINSPACE
# =========================================

print("\nLinspace:")
print(np.linspace(1., 5., num=20))

# =========================================
# 16. TRUY CẬP PHẦN TỬ
# =========================================

print("\nFirst Row:")
print(np_arr[0])

print("\nFirst Column:")
print(np_arr[:, 0])
```

```python
import numpy as np
import pandas as pd

# =========================================
# 1. SERIES
# =========================================

a = pd.Series([1, 2, 3], index=['a', 'b', 'c'])

print("Series a:")
print(a)

# Truy cập phần tử
print("\nAccess a['a']:")
print(a['a'])

# =========================================
# 2. SERIES TỪ DICTIONARY
# =========================================

week = {
    "t2": "thuhai",
    "t3": "thuba"
}

b = pd.Series(week)

print("\nSeries từ Dictionary:")
print(b)

# =========================================
# 3. DATAFRAME TỪ NUMPY ARRAY
# =========================================

c = np.array([
    [1, 2],
    [3, 4]
])

df_c = pd.DataFrame(c)

print("\nDataFrame từ NumPy Array:")
print(df_c)

# Chuyển DataFrame -> NumPy Array
print("\nDataFrame -> NumPy Array:")
print(np.array(df_c))

# =========================================
# 4. DATAFRAME TỪ DICTIONARY
# =========================================

dic = {
    'name': ['nguyen van a', 'nguyen van b'],
    'age': [10, 20]
}

print("\nDictionary:")
print(dic)

df_dic = pd.DataFrame(dic)

print("\nDataFrame từ Dictionary:")
print(df_dic)

# =========================================
# 5. RANDOM DATAFRAME
# =========================================

# randn:
# random số thực theo phân phối chuẩn

d = np.random.randn(20, 2)

df_d = pd.DataFrame(
    d,
    columns=['so1', 'so2']
)

print("\nRandom DataFrame:")
print(df_d)

# =========================================
# 6. THÔNG TIN DATAFRAME
# =========================================

print("\nShape:")
print(df_d.shape)

print("\nColumns:")
print(df_d.columns)

print("\nDtypes:")
print(df_d.dtypes)

# =========================================
# 7. TRUY CẬP CỘT
# =========================================

print("\nColumn so1:")
print(df_d['so1'])

# =========================================
# 8. TRUY CẬP NHIỀU CỘT
# =========================================

print("\nColumns so1 và so2:")
print(df_d[['so1', 'so2']])

# =========================================
# 9. LẤY DÒNG ĐẦU
# =========================================

print("\nFirst 5 rows:")
print(df_d.head())

# =========================================
# 10. THỐNG KÊ CƠ BẢN
# =========================================

print("\nDescribe:")
print(df_d.describe())
```

```python
import tensorflow as tf
import numpy as np

rank_0_tensor = tf.constant(4);
print(rank_0_tensor);

print(tf.constant(9, tf.int16))
print(tf.constant(1.12, tf.float32))

print(tf.constant([1.0, 2.3, 3.4]))

print(tf.constant([[1, 2],[3, 4],[5, 6]], tf.int16))

print(tf.constant(np.random.randint(2,4), dtype=tf.int64))

print(tf.fill(3, 10))

print(tf.fill([3,3], 9))

print(tf.linspace(5, 20, 5))

print(tf.range(3, 7, delta=0.7))

print(tf.random.normal([5]));

print(tf.random.normal([3,3]))

print(tf.random.truncated_normal([5]))

print(tf.random.truncated_normal([3,2,4]))

print(tf.random.uniform([3, 2], minval=10, maxval=20))

df_e = tf.random.uniform([5], minval=1, maxval=20)

print(df_e)

print(tf.random.shuffle(df_e))

print(tf.cast(df_e, dtype=tf.int32))

print(tf.sqrt(df_e))

print(df_e + tf.sqrt(df_e))

print(tf.add(df_e,tf.sqrt(df_e)))

print(df_e * tf.sqrt(df_e))

print(tf.multiply(df_e,tf.sqrt(df_e)))

print(tf.subtract(df_e,tf.sqrt(df_e)))

print(tf.divide(df_e,tf.sqrt(df_e)))

print(tf.scalar_mul(3, df_e))

print(tf.math.reciprocal(df_e))

print(tf.math.mod(df_e, tf.sqrt(df_e)))

print(tf.negative)

print(tf.round(df_e))
print(tf.math.rint(df_e))
print(tf.math.ceil(df_e))
print(tf.math.floor(df_e))












```
